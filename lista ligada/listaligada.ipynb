# Listas Ligadas

São uma estrutura simples com grande diferencial em acessar o meio da lista em O(n), facilitando inserção e deleção no meio da lista (principalmente se já estiver no nó). É mais vantajoso para a escrita, mas pode ser mais complexo para a leitura

Apesar de usar mais memória por conta dos ponteiros, podem crescer dinâmicamente, sem precisar alocar um bloco inteiro de memória logo que é criado.

### Como criar

pode ser uma lista circular, onde o último nó sempre leva ao primeiro, uma lista que aponta só para frente ou uma lista que aponta pra frente e pra trás

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class linkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def append(self, value):
        if not self.head:
            self.head = Node(value, None)
            self.tail = self.head
        else:
            new_node = Node(value)
            self.tail.next = new_node
            self.tail = new_node
    
    def print_list(self):
        current = self.head
        while current:
            print(current.value, end=" -> ")
            current = current.next

    def find(self, value):
        current = self.head
        counter = 0
        while current:
            counter += 1
            if current.value == value:
                return counter
            current = current.next
        return False
    

    #essa função retorna o primeiro nó dos meios caso seja par
    def middle(self):
        slow = self.head
        fast = self.head

        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
        
        return slow.value if slow else None
```
